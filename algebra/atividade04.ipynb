### Problema 1

Qual é a **forma fatorada** do produto entre os polinômios  
$$
x^2 + 14x + 49
\quad\text{e}\quad
x^2 - 14x + 49
$$

A) $(x + 7)^2 \cdot (x - 7)^2$  
B) $(x^2 + 14x + 49) \cdot (x^2 - 14x + 49)$  
C) $(x + 7) \cdot (x - 7)^2$  
D) $(x + 7)^2 \cdot x - 7^2$  
E) $x + 7^2 \cdot (x - 7)^2$
